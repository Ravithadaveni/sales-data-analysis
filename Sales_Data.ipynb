##  Step 1: Load the Dataset

```python
import pandas as pd

# Load the uploaded dataset
file_path = r"C:\Users\HP\OneDrive\Desktop\Data_Sets_kaggale\Sales_Data.csv"
df = pd.read_csv(file_path)
```

## Step 2: Data Cleaning  
**Check for null values**  
- Handle duplicates  
- Understand the types of data

```python
# Check for missing values
print(df.isnull().sum())

# Remove duplicates if any
df.drop_duplicates(inplace=True)

# Optional: Fill or drop nulls
df.fillna(0, inplace=True)  # Example
```

## Step 3: Exploratory Data Analysis (EDA)  
- Basic stats & insights:

```python
# Summary statistics
print(df.describe())

# Value counts for categorical columns
print(df['column_name'].value_counts())  # Replace column_name
```

## Step 4: Visualizations  
- You'll use Matplotlib and Seaborn:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Sales distribution
sns.histplot(df['Sales'])  # Replace 'Sales' with correct column
plt.title("Sales Distribution")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include='number')  # Select only numeric columns
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

##  Key Insights

-  Most sales are concentrated between 200 and 500, showing that mid-range products are most common.
-  The 'Supermarket Type1' outlet has the highest number of transactions and total sales.
-  'Fruits and Vegetables' and 'Snack Foods' are among the top-selling product categories.
-  Theres a weak correlation between item MRP and item visibility  which might indicate that visibility doesnt influence pricing.
-  Standard Class shipping is the most used method, but might need optimization if sales or delays are an issue.
